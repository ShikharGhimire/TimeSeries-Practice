- A common process of time series is to create data based off a rolling mean. What we can do is divide the data into windows of time and then calculate aggregrate function for each window.
- This way we obtain a simple moving average

```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/shikharghimire/Desktop/TimeSeries-Practice/data/starbucks.csv',index_col = 'Date',parse_dates = True)
```

```python
df.head()
```

```python
df['Close'].plot(figsize=(12,5))
```

```python
#Add in the rolling mean
#What rolling mean does is create a little window. You can choose that window. Let's say if it is 7 days window. 
#It then looks at each section of 7 days and perform some sort of aggregate function on it

#Let's plot it

df['Close'].plot(figsize=(12,5))
df.rolling(window=7).mean()['Close'].plot()

#The. first six values are all nan because at first 6 values, we haven't reached 7 days yet to take the mean
```

```python
#We can create a different coloumn for let's say 7 days rolling window mean
df['Close : 7 days mean'] = df['Close'].rolling(window=30).mean()
```

```python
#We can also plot two coloumn at the same time
df[['Close','Close : 7 days mean']].plot(figsize = (12,5))
```

```python
#Let's look at expanding now
#Instead of calculating values for rolling window updates, let's take count of everything from start of the time series
#upto each point of time. Instead of considering average over the. last 7 days, we. will consider. all prior data in our expanding
#set of averages
df['Close'].expanding().mean().plot(figsize = (12,5))

#essentially what it is showing is the expanding windows. Instead of shifting windows by 7 days or 30 days, what we 
#do is at each time point is to take the average from the previous point 
```
